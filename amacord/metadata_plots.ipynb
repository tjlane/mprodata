```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
metadata = pd.read_csv("unique_mpro_datasets_prior_to_downselection.csv")
```

```python
creators = metadata["creator_id"].unique()

fig, axes = plt.subplots(len(creators)//3, 3, sharex=True, figsize=(6,18))

for ic, creator in enumerate(creators):

    ax = axes[ic//3, ic%3]

    subset = metadata[metadata["creator_id"] == creator]
    ax.hist(subset["volume"], bins=np.linspace(250,270,21), alpha=0.5, label=creator)
    ax.set_title(creator)

plt.tight_layout()
```
